Evrim Ağacı youtube kanalının videolarını analiz ederek oluşturduğum veri setine k-means uygulanması.

```python
import pandas as pd

video_analizi= pd.read_csv('C:/Users/Asus/video_analiz/video_analiz2.csv',sep=";")
video_analizi
```

```python
video_analizi.columns
```

```python
import matplotlib.pyplot as plt

plt.scatter(video_analizi['Begeni'], video_analizi['Yorum'])
plt.title('Evrim Ağacı Videolarının Analizi')
plt.xlabel('Beğeni Sayısı')
plt.ylabel('Yorum Sayısı')
plt.show()
```

```python
import os
os.environ["OMP_NUM_THREADS"] = "1"

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, n_init=10,random_state=42)
kmeans.fit(video_analizi[['Begeni', 'Yorum']])
video_analizi['Küme'] = kmeans.labels_
video_analizi
```

```python
kmeans.cluster_centers_
```

```python
import matplotlib.pyplot as plt

plt.scatter(video_analizi['Begeni'], video_analizi['Yorum'], c=video_analizi['Küme'])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red')
plt.title('Evrim Ağacı Videolarının Analizi')
plt.xlabel('Beğeni Sayısı')
plt.ylabel('Yorum Sayısı')
plt.show()
```

```python
#standardizasyon yapalım.Beğeni sayısını 15'e bölelim.
video_analizi['Begeni'] = video_analizi['Begeni'] /15
video_analizi
```

```python
import matplotlib.pyplot as plt

plt.scatter(video_analizi['Begeni'], video_analizi['Yorum'])
plt.title('Evrim Ağacı Videolarının Analizi')
plt.xlabel("Beğeni Sayısı(15'e bölünmüştür)")
plt.ylabel('Yorum Sayısı')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(video_analizi['Begeni'], video_analizi['Yorum'], c=video_analizi['Küme'])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red')
plt.title('Evrim Ağacı Videolarının Analizi')
plt.xlabel("Beğeni Sayısı(15'e bölünmüştür)")
plt.ylabel('Yorum Sayısı')
plt.show()
```
